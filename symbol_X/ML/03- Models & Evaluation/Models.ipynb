#Import

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
```

#Data preparation

```python
assets = 'assets.csv'

assets = pd.read_csv(assets)

y = assets.Price

assets_features = ['Market Cap (in M)', 'Beta', 'Volume 52 weeks', 'Performance (52 weeks)', 'RÃ©sultat net', 'Price 52 Weeks Ago', 'Total assets']

X = assets[assets_features]

train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 1)
```

#Linear Regression

```python
assets_model = LinearRegression()

assets_model.fit(train_X, train_y)

val_predictions = assets_model.predict(val_X)

print('Val prediction :', val_predictions)

print('MAE :', mean_absolute_error(val_y, val_predictions))
```

#Decision Tree

```python
assets_model = DecisionTreeRegressor()

assets_model.fit(train_X, train_y)

val_predictions = assets_model.predict(val_X)

print('Val prediction :', val_predictions)

print('MAE :',mean_absolute_error(val_y, val_predictions))
```

#Random Forest Regression




```python
assets_model = RandomForestRegressor()

assets_model.fit(train_X, train_y)

val_predictions = assets_model.predict(val_X)

print('Val prediction :', val_predictions)

print('MAE :',mean_absolute_error(val_y, val_predictions))
```
