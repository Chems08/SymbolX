# Modules

```python
import pandas as pd
```

##Load data

```python
df = pd.read_csv('assets.csv')
```

#Version 1

*   Step 1 : Drop lines with missing values
*   Step 2 : Drop duplicates


```python
#Etape 1
df = df.dropna()

#Etape 2
df = df.drop_duplicates()

df.to_csv('cleaned_data.csv', index=False)
```

#Version 2


*   Step 1 -
Drop duplicates

*   Step 2 -
Drop some columns

*   Step 3 -
Drop lines that has missing values in some columns

*   Step 4 -
Drop lines that has 0 in some columns

```python
#Etape 1
df = df.drop_duplicates()

#Etape 2
df = df.drop(columns=["Chiffre d'affaires", 'Dividend Per Share Annual', 'EBITDA CAGR (5y)', 'EBITDA', 'Dividend Yield Indicated Annual', 'P/E Ratio', 'Ratio Debt/Equity (Annual)'])

#Etape 3
df = df.dropna(subset=['Symbol', 'Beta', 'Sector', 'Industry', 'Currency', 'Performance (52 weeks)', 'Price 52 Weeks Ago', 'Résultat net', 'EPS Annual', 'ROI Annual', 'Total assets', 'Volume 1 month', 'Exchange'])

#Etape 4
for column in ['Market Cap (in M)', 'Volume 52 weeks', 'EPS Annual', 'ROI Annual', 'Volume 1 month']:
    df = df[df[column] != 0]

df.to_csv('cleaned_asset_data.csv', index=False)
```
